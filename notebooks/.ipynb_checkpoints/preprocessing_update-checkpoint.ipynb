```python
# load spellcheck file
automatic_suggestions = pickle.load(open("../automatic_suggestion.pkl","rb"))
bad_suggestions = pickle.load(open("../bad_suggestion.pkl","rb"))
own_corrections = pickle.load(open("../own_correction.pkl","rb"))
```

```python
airlines = pd.read_csv("../data/openflights/data/airlines.csv",sep = ',',header = None)
airlines.rename({1:'name'}, axis = 1, inplace = True)
airline_names_list = list(airlines["name"].str.lower())

def change_spelling(texts,misspelled_lists,ignore_list,airlines,
                    automatic_suggestion = None, bad_suggestion = None, own_correction = None):
    if automatic_suggestion is None:
        automatic_suggestion = []
        bad_suggestion = []
        own_correction = {}
    for index, text in enumerate(misspelled_lists):
        print("processing text:", index)
        text_or = " ".join(texts[index])
        for word in text:
            if word not in ignore_list:
                if word in airlines:
                    print("word:",word,"identified as airline")
                    text_or = re.sub(word, "airline", text_or)
                else:
                    if word in automatic_suggestion:
                        print("automatic spelling correction")
                        text_or = re.sub(word, spell.correction(word), text_or)
                    elif word in list(own_correction.keys()):
                        text_or = re.sub(word, own_correction[word], text_or)
                    elif word not in bad_suggestion:
                        if word == spell.correction(word):
                            automatic_suggestion.append(word)
                        else:
                            print("original word:", word)
                            print("suggestion is:", spell.correction(word))
                            print("type ok if it is a good suggestion or 'change' if you want to rewrite it ")
                            ok = input()
                            if ok == 'ok':
                                text_or = re.sub(word, spell.correction(word), text_or)
                                automatic_suggestion.append(word)
                            elif ok =='change':
                                print("type in your change:")
                                change = input()
                                text_or = re.sub(word, change, text_or)
                                own_correction[word] = change
                            else:
                                bad_suggestion.append(word)
        texts[index] = text_or
        #if index == 5:
        #    break

    return texts, automatic_suggestion, bad_suggestion, own_correction

texts, automatic_suggestions, bad_suggestion, own_correction = change_spelling(texts = texts,
                                                               misspelled_lists = misspelled,
                                                               ignore_list = ["google", "upload", "website", 
                                                                              "layover","bogot", "inbox", "rebook", 
                                                                              "checkin","plugins","facebook",
                                                                              "paypal"],
                                                               airlines = airline_names_list,
                                                               automatic_suggestion = automatic_suggestions,
                                                               bad_suggestion = bad_suggestions,
                                                               own_correction = own_corrections)
```

```python
def check_cities(texts_array,changes = None, ignore_list = None):
    texts = texts_array.copy()
    if not changes:
        changes = []
    if not ignore_list:
        ignore_list = []
    for i in range(len(texts)):
        print("text", i)
        for change in range(len(changes)):
            changed_text = re.sub(changes[change],'city',texts[i])
            if changed_text != texts[i]:
                texts[i] = changed_text
                print(changes[change],"was changed to city")
        doc = nlp(texts[i])
        for ent in doc.ents:
            if ent.label_ == 'GPE':
                print(changes)
                if (ent.text in changes) or (ent.text in ignore_list):
                    continue
                print(ent.label_, ent.text)
                response = input()
                if response == 'ok':
                    texts[i] = re.sub(ent.text,'city',texts[i])
                    changes.append(ent.text)
                elif response == 'change':
                    city_name = input()
                    texts[i] = re.sub(city_name,'city',texts[i])
                    if city_name not in changes:
                        changes.append(city_name)
                else:
                    ignore_list.append(ent.text)
    
    return texts, changes, ignore_list

edited_texts, changes, ignore = check_cities(texts,changes,ignore)
```

```python
#BASE VERSION NOTHING, USE ONLY LEMMATIZATION IN VERSION 2
# Lemmatize with POS Tag
from nltk.corpus import wordnet
nltk.download('averaged_perceptron_tagger')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

# 1. Init Lemmatizer
lemmatizer = WordNetLemmatizer()
#stop_words = set(stopwords.words('english')) 
stop_words = []
# 2. Lemmatize a Sentence with the appropriate POS tag
for i in range(0,len(edited_texts)):
    sentence = texts[i]
    word_tokens = nltk.word_tokenize(texts[i])
    filtered_sentence = [w for w in word_tokens if not w in stop_words] 
    #print(filtered_sentence)
    edited_texts[i] = ' '.join([lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in filtered_sentence])
                         
```
